```python
# Importa as bibliotecas
import pandas as pd
import os
```

```python
# Verifica a pasta corrente
pasta = os.getcwd()
```

```python
pasta
```

```python
# Lista arquivos da pasta
arquivos = os.listdir(pasta)
```

```python
arquivos
```

```python
# Lista os arquivos somente excel e que comecem com 'ambos'
arquivos_xls = [arq for arq in arquivos if arq[-3:]=='xls' and arq[0:5]=='ambos']
```

```python
type(arquivos_xls)
```

```python
arquivos_xls
```

```python
len(arquivos_xls)
```

```python
# Inicializa um dataframe vazio
df_ambos = pd.DataFrame()
```

```python
pular_linhas = list(range(0,5)) + list(range(46,62)) + list(range(102,113))
for arq in arquivos_xls:
    dados=pd.read_excel(arq, 
                        skiprows = pular_linhas,
                        usecols = "A:G")
    df_ambos = df_ambos.append(dados)
```
